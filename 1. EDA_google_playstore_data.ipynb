```python
#1 MB > 1024KB


#method1
def size_mapping(item):
    if str(item)[-1] == 'M':
        return float(str(item).replace("M", "")) * 1024
    elif str(item)[-1] == 'K':
        return float(str(item).replace("K", ""))
    else:
        return str(np.NaN)
```

```python
float(str('50M').replace("M", "")) * 1024
```

```python
size_mapping("50K")
```

```python
size_mapping("varies with device")
```

```python
df['Size'].apply(size_mapping)
```

```python
#use regex to do the above functionality
def process_size(series):
    return series.str.extract(r'([\d\.]+)([MmKk])').apply(
        lambda row: float(row[0]) * 1024 if str(row[1]).lower() == 'm'
        else float(row[0]) if str(row[1]).lower() == 'k'
        else np.nan, axis=1
    )
```

```python
df['Size'].str.extract(r'([\d\.]+)([MmKk])')
```

```python
process_size(df['Size'])
```

```python
#another method:

def process_sizes(series):
    series = series.str.lower()
    m_mask = series.str.endswith("m")
    k_mask = series.str.endswith("k")
    numeric = series.str.extract(r"([\d\.]+)").astype(float)

    numeric[m_mask] *= 1024
    # k_mask values remain as is
    return numeric[0]


process_sizes(df['Size'])
#this method will be preferred as its a vectorized method
```

```python
df['Size'] = process_sizes(df['Size'])
```

```python
#cleaning steps 
#4.0.3  becomes 4.0
#4.4W >> 4.4
#'5.5-8.9' >> 5.5
```

```python
def clean_android_version(val):
    if pd.isna(val) or val == "":
        return np.nan
    # Remove whitespace
    val = val.strip()
    # Handle version ranges: keep min version (before '-')
    if '-' in val:
        val = val.split('-')[0].strip()
    # Handle weird suffixes like '4.4W' >>'4.4'
    val = ''.join([ch for ch in val if ch.isdigit() or ch == '.'])
    return val if val else np.nan
```

```python
df['Android Ver'] = df['Android Ver'].apply(clean_android_version)
```
